```python
import pandas as pd
import numpy as np
```

```python
car_sales = pd.read_csv('7.1 car-sales.csv')
```

```python
car_sales
```

```python
car_sales["Price"]= car_sales["Price"].str.replace('[\$\,\.]', '').str[:-2].astype(int)
```

```python
car_sales
```

```python
car_sales["Date"]= pd.date_range("1/1/2020", periods= len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(y= "Date", x ="Total", kind = "scatter");
```
